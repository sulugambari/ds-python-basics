
Python always has more to offer. 
As well as simple operators you can find on any calculator, Python has two more arithmetic operators which will be of use to you: 

The double forward slash `//` is called **floor division**. All it does is perform division and truncate the result to the nearest whole number.

````python
# Floor division
34 // 4
# 34 / 4 = 8 remainder 2
````

The **modular division** operator, `%` (also called **modulo**, "percentage symbol"). 
This operation is the sibling to `//`.
**BUT**, whereas `//` returns the truncated result of a division, `%` returns the remainder of integer division.

````python
# Modulo
34 % 4
````

## Variables

One of the most powerful constructs in programming is the ability to store arbitrary values in what are called **Variables**.
You can think of variable assignments as giving a name to something, so that it can be accessed later by different parts of your program.

Variable assignment in Python occurs with the `=` operator. To assign a value to a variable name (i.e. to **declare** it as a variable), simply put the variable name on the left side of the `=` and the value you want to associate with that variable name on the right side. 
After it is declared, you can immidiately access the value in the variable using it's name somewhere later in the code or notebook.

````python
a = 8
b = "Hello "
````

````python
b
````

````python
a - 4
````

````python
a * b
````

### Naming Variables

The name given to a variable can technically be any contiguous set of characters, but there are some naming conventions followed in Python and programming in general. 

Python follows a variable naming convention called **snake_case**. snake_case uses underscores `_` instead of spaces and only lower case letters (e.g. `my_first_variable`). 

Although `my_first_variable`is in snake_case it is not a good variable name. In order to make your code more readable, and maintainable you should always give your variables **well-defined**, **succinct** names. Calling something `height` is a lot more meaningful than calling it `x`. 

Of course, there are cases when using a less than descriptive variable name follows convention, thats fine to do. A common example is the use of `i` to keep track of an **index**, because of its prevalent usage for indexing, it is usually easy to understand what is happening in that context when all you see is the variable named `i`.

### Calling and Changing Variables

In the cells above `a = 8` or `b = "Hello"` did not generate an output. This is because the return value that would have been printed as output was assigned to the variables `a` and `b`. To view the content of the variables we had to call them in the next lines.

A large part of a variables' **power** is the fact that they can change. This allows us to use a single variable name to keep track of a specific thing throughout the life of a program. 
Remember how we assigned the value `8` to variable `a` above? The exact same syntax can be used to change the value stored in the variable.
Let's say we want to increase the value of variable `a` by 5?

````python
a = a + 5
a
````

Notice how the first line above is formatted. Python knows that the `=` means variable assignment, so when it sees the first line it evaluates the right side of the equals and then puts that value in `a`, even though `a` is part of the calculation on the right side. `a` is now connected with this new value and the old value is gone.

Changing variables in this way occurs so commonly that there is built-in **shorthand** for it. The result of the first line could have been achieved with `a += 5`. This *syntactic sugar* is available for all simple operations `+`, `-`, `*`, `/`, `**`, and `%` that are covered above.

## Types

Data types are an important concept in programming. 
Variables can store data in different types and different data types do different things.

Remember the adage: "If it walks like a duck, and quacks like a duck, then it must be a duck". 
Python is a **duck-typed** language, meaning that Python will determine what it thinks is the best type to call a variable when used, unless explicitly told otherwise.

### Numeric Variable Types

Let's review some basic numeric variable types in Python, which all represent numbers in some way:

The most important ones are:

* `int`, short for integer(s), whole numbers without a decimal point

* `float`, short for floating point/decimal number(s), numbers with a decimal point

* `complex`, which contain real and imaginary parts stored as floats

To inspect what type Python thinks a variable (or number...) is, you can use the `type()` function. 
This will be especially useful when running into an error and will help to debug code. 

````python
# Int 14 as variable c
c = 14
````

````python
type(c)
````

````python
# Float 32.2 as variable d
d = 32.2
````

````python
type(d)
````

````python
# Complex(5, 6) as variable e
e = complex(5, 6)
````

````python
type(e)
````

Python assumes that numbers with no decimal points are `int`, those with decimal points are `float`, and (unsurprisingly) those from the `complex()` constructor as `complex`.
Frequently, these subtle differences don't matter too much. However, there will be times when the implementation detail will make you think that something will work, when really it won't. Knowing how to check the **type()** of something will help solve any of these potential problems.
