So far, we have iterated through lists, which are ordered structures. These are great containers if there is some intrinsic order to the data that we're storing. 

However, there are plenty of times when we don't care about order, either because it simply doesn't matter or because the data are associated with each other in different ways. For example, say we have a few American state names, and we want to associate each state's name with its capital. How would we do this in a list? 

One way would be to have tuples that store pairs of states and their capitals.

````python
states_caps = [
    ("Georgia", "Atlanta"),
    ("Colorado", "Denver"),
    ("Indiana", "Indianapolis"),
]
states_caps
````

As simple as it looks, there are limits to how intuitive this storage method is. 

Consider that if we wanted to find the capital of Indiana (Indianapolis), we would have to search through the entire list, checking to see if Indiana is in the first position of each tuple. If/when we find it, we would then have to grab the second position of that tuple

Here is an example of how this could be done. 

````python
search_state = "Bavaria"
capital = "State not found"  # Default result if the state is not found in the list

for state_cap in states_caps:
    if state_cap[0] == search_state.capitalize():
        capital = state_cap[1]
        break
capital
````

While this looks okay, we can do much better. Python **Dictionary** to the rescue!

The dictionary data structure in Python allows us to store data in a way that is more **intuitive** for this problem. 

Dictionaries allow us to store a value associated with a **Keyword**. In the example above, we want to store the capital as the **Value**, and the state as the **Keyword** that the capital is associated with. There are many ways to instantiate a dictionary. Let's look at the **simplest**.

````python
empty_dict = {}  # This is how you create an empty dictionary. Looks Familiar doesnt it?

states_caps_dict = {
    "Georgia": "Atlanta",
    "Colorado": "Denver",
    "Indiana": "Indianapolis",
}
states_caps_dict
````

This looks very similar to the way that we made lists and tuples, except now we use curly braces, and there is the new use of colons `:`. 

On the left side of each colon we have the **Keyword**, and on the right side of the colon we have the **Value** associated with it. Each **Key-Value** pair, as we call them, is then separated by a comma.

It is also possible to use the "dict()" constructor to make a dictionary. 

````python
dict(states_caps)
````

This can be done in a few ways. 

````python
empty_dict = dict()  # Creates an empty dictionary that can be filled later
list_to_dict = dict(
    [(1, "one"), (2, "two"), (3, "three")]
)  # Converts a list of tuples into a dictionary
kwargs_to_dict = dict(
    Age=30, Height=180, Weight=80
)  # Creates a dictionary using keyword arguments

print(empty_dict, type(empty_dict))
print(list_to_dict, type(list_to_dict))
print(kwargs_to_dict, type(kwargs_to_dict))
````

How do we use dictionaries? Let's take an example from above and say we're trying to find the capital of **Atlanta**. 

With a list of tuples, we would have to:

1. Iteratively search through the list of tuples from the beginning to find the one with 'Atlanta' in the first position
2. Grab the second entry in that tuple to return the capital city. 

With dictionaries it's much easier. Run the cells below to find out.

````python
states_caps_dict["Georgia"]
````

````python
# Intentional error example kept for the exercise.
# Uncomment the lines below to see the error.
# It should raise KeyError.
#
# states_caps_dict["Washington"]
````

As you can see, all we have to do is **index** into the dictionary, like we did with lists, but this time using the **key** instead of the index. The dictionary then returns the associated value. 

**Notice**: In the second cell we tried to find a key that wasn't already in the dictionary. This returned a `KeyError` telling us that that key is not stored in the dictionary.

This error message shouldn't happen too frequently because we often know the keys in our dictionaries. But, for times where we don't know if a key is already in the dictionary, we luckily have the `get()` method. 

The `get()` method takes tewo arguements. 

1. The key you're trying to find
2. A default **return value** that is returned if the key doesn't exist

Now, lets try it out.

````python
states_caps_dict.get("Indiana", "State not found")
````

````python
states_caps_dict.get("Washington", "State not found")
````

By using the `get()`method on our dictionary, we asked if the `states_caps_dict` for the value associated with the keys `'Ìndiana'` and `'Washington'`, and told it to return `'State not found'` if the keyword wasn't in the dictionary. 

For `'Washington''` we get back `'State not found'`. This makes sense because we know that `'Washington'` is not in the dictionary.

## Mutability of Dictionaries

Are dictionaries mutable? That's a great question! Yes they are! 

Before we talk about how to "mutate" dictionaries, let's describe them in the same language that we used for lists and tuples. A dictionary is defined as an unordered collection of key-value pairs, where each key is required to be **Unique**.

With that in mind, let's recall how we "mutated" a **list**. 

To change an element at an existing index, we indexed into the list and did assignment. To make them larger, we can use the `append()` and `insert()` methods. These methods of mutation make a lot of sense, considering that lists are ordered. 

**But**, dictionaries live in the **unordered** data paradigm, so to change a key-value pair, all you have to do is index into it with the existing key and assign a value to it. And, to add a new key-value pair, declare a new key and assign it a value. **Notice** how assignment works just as before, with the `=`. 

Let's take a look.

````python
my_dict = {"Thing": "One", "Other": "Two"}
my_dict["Thing"]
````

````python
my_dict = {"Thing": "One", "Other": "Two"}
my_dict["Other"]
````

````python
my_dict["Thing"] = "Three"
my_dict
````

````python
my_dict["Other Things"] = "Four"
my_dict
````

As we have seen, Dictionaries make it easy to store key-value relationships in a single data structure. They are also designed for easy value retrieval, which would otherwise be longer if the information were to be stored in a list. 

What are the restrictions on the data types you can put in a dictionary? 

As for the values, like in lists, there are none! But in terms of the keys associated with those values, that's a different story. **Keys** in dictionaries **must** be an **immutable** type, and if that type is a container, then the container cannot contain any mutable types. 

Why is this the case? The answer lies in the way that dictionaries store values and associate them with a key. Python dictionaries are an implementation of what's known as a *hash map* or *hash table* ([Here's](https://en.wikipedia.org/wiki/Hash_table) the Wikipedia page for them to geek out on). This computer science idea is basically a function that relates any input, in our case the **keys**, to a location in memory. 

Thus, retrieval of a value from a dictionary is entirely dependent on the key. If we were to use a mutable type as the key for a dictionary and later change what that key looked like by mutating it, the dictionary wouldn't be able to find the value it was supposed to associate with that key (as the key has now changed). 

The code below will **not** run. It is there to demonstrate what the type of incorrect key usage would look like.


```python
my_bad_key = ['key']    # Original key
my_dict = {my_bad_key: 'This wont work'}    # Dictionary declared with a list as a key (won't work)
my_bad_key.append('other_key')  # Let's append to our mutable 'key'
my_dict[my_bad_key]     # How is the dictionary supposed to know what we're looking for???
```

### Getting More Out of Dictionaries

We now know how to make and alter dictionaries, and how to use them to store arbitrary key-value pairs. Now, let's talk about how to use them with loops.

As with lists and tuples, dictionaries are iterables in Python. This means that Python knows how to traverse everything that's stored in the collection. The way we did this with list was with a `for` loop. We will again use the `for` loop with dictionaries. 

**Note**: There are a few changes in how a `for` loop is implemented, since dictionaries are unordered key-value pairs, whereas lists are ordered collections of values. 

Let's revisit below how we traverse a list with a `for` loop.

````python
for element in list(range(10)):
    if element % 2 == 0:
        print(element)
````

In each iteration of the `for` loop, we grab a number from the list one at a time, give it the name `element`, check if it's even, and then print that value if it is. 

We want to call you're attention to the phrase **one at a time**. Lists are an ordered collection of values; dictionaries, on the other hand, have keys and values that are tied together. Therefore, iterating into a dictionary with a `for` loop, you can expect to only get one of these out (keys or values). Naturally, it's the dictionary keys.

````python
for element in states_caps_dict:
    print(element)
````

This can also be achieved using the `keys()` method, which explicitly returns the keys of the dictionary.

````python
for key in states_caps_dict.keys():
    print(key)
````

**Remember** that dictionaries are unordered. So when we access the keys, we are not guaranteed any particular order when accessing a dictionary's keys. Thus they may not be printed in order.

So we can loop through all the keys, and the natural question that follows this is whether we can loop through all of the values? 

This can be done with the `values()` method.

````python
for value in states_caps_dict.values():
    print(value)
````

This is a very useful feature, but it gets better! 

One of the most useful ways to loop through the contents of a dictionary is by getting each key-value pair together within the loop. The `items()` method does exactly this. To use it, we will employ a similar syntax to what we used with `enumerate()` list method.

````python
for state, capital in states_caps_dict.items():
    print(state, capital)
````

As a brief learning tangent, let's discuss what's happening when we use this syntax.

 As above, we are going to use the `items()` method, but this time not store the output as individual `state` *and* `capital` variables.

````python
for thing in states_caps_dict.items():
    print(thing)
````

Now that we're only using a single variable to grab the output of the `items()` method, we can see that the method is outputting a tuple for each key-value pair. 

What was happening when we used `state` and `capital` to grab the output?? Very frequently, we want to put the separate values from ordered collections into different variables. This happens so frequently, in fact, that Python has a built-in way to do it quickly called **unpacking**.

When Python sees the two variable names `state` and `capital` in the first implementation, it automatically knows to take the values in the tuple returned from `items()` method and put the first one in `state` and the second in `capital`. This is also what was happening when we called `enumerate()` on a list - it returned a tuple with the index as well as the value itself. 

It is up to you whether or not to grab those values in a single variable as a tuple, or have Python unpack it for you into two variables.

**Note**: Python will not allow you to "unpack" a collection containing a single item into multiple variables.
