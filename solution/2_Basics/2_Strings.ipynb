## Strings

First, lets learn about another common data type, **strings**. 
From a high-level perspective, a string is just a piece of text. This could be text read in from a file, html pulled from the Internet, or any other text. From Python's perspective, a string (type(`str`)) is simply a collection of **encoded** characters. 

Wait, what's an **encoding**? An encoding is just a fancy way of saying that the characters in a string follow a certain **format**, or **structure**. 

The reason this matters is due to the `terms` of Python programs. Python expects strings to be in one of a couple of different encodings (either `ASCII`, `utf-8`, or `unicode`). This isn't something you will often run into, and especially not when defining strings (it will probably be most prevalent when pulling text from the internet). However, it is worth mentioning here because there is a good chance that somewhere down the line the Python compiler will tell you it doesn't recognize a certain character in a string or strings, and an unexpected encoding will most likely be at the heart of that error.

In Python, strings are recognized as a collection of characters surrounded by a set of either single quotation marks (`'...'`) or double quotation marks (`"..."`). you are free to use either **single** or **double** quotation marks, so long as they open and close the string with a **matching** set. The single caveat to that is that if you are writing an expression with a single quotation mark in it (such as "Don't do that"), you will **have to** use a matching set of **double** quotation marks. 

Let's experiment with some strings...

````python
f = "This is a string."
````

````python
f
````

````python
type(f)
````

````python
g = "This is also a string, but this time with double quotation marks."
````

An error occurs, why? 

Well, the string is opened with a single `quotation` mark, so the Python compiler started searching for the next single quotation mark to close the string. When it found the quotation mark in the word **didn't**, it assumed the string was closed after `didn`.
As a result, this left `t listen to them.` just hanging out on its own. The compiler didn't know how to interpret that, resulting in SyntaxError.
The solution to this, as mentioned above, is to use double quotation marks in any case where your text will likely have single quotation marks in it. 

For example...

````python
"Now that I've got double quotes, I'm free to use all the contractions and/or use quotes etc..."
````

````python
"Can't, won't, didn't, don't... all the contractions! and even many ''''''''''''''''''"
````

### String Operations

Surprisingly, a couple of standard mathematical operations will work on strings, namely `+` and `*`. We can use the `+` operator to add two strings together (this is known as String **Concatenation**), and we can use the `*` operator to repeat a string a given number of times. 

Let's take a look...

````python
"My first string" + "My second string"
````

````python
"Repeating string" * 3
````

Notice that Python didn't put spaces between the strings with either the `+` operator or the `*` operator. 

Why not? 

Simply, because it wasn't told to! In this case and in programming generally, We always have to be extremely explicit about what we want the computer to do. To fix this, we can add a space in the middle of the first case, and then add a space to the end of our string in the second case. This can be done in a variety of ways.

````python
"My first string " + "My second string"  # Space after the first string
````

````python
"My first string" + " My second string"  # Space before the second string
````

````python
"My first string" + " " + "My second string"  # Special single spaced third string
````

````python
"Repeating string " * 3
````

````python
("Repeating string" + " ") * 3
````

That looks much better! 

But, what about the pesky little space at the end of the string: `'Repeating string Repeating string Repeating string '`... 
How can we remove this? 

It turns out there is a special string **Method** - A name for a **function** that is attached to a **particular** **object** **type**.

methods are something that we will cover in much more depth later, but for now just note that **dot notation** is used to call them on our objects. 

**Dot Notation** - Placing a full stop `.` at the end of the object (`str`, `int`, `float` or any variable) to call a specific object method. 

Here is how the use **dot notation** for the `strip()` method.  

````python
"Repeating string Repeating string Repeating string ".strip()
````

````python
" Repeating string Repeating string Repeating string ".strip()
````

In the first example above, the moethod removed the trailing space from the string. In the second example, it removed both the leading and trailing spaces. This is exactly what the `strip()` method does - by default (without any arguments) it removes leading and trailing whitespace.

*Note: **The method can actually remove any leading or trailing characters if you pass them to `strip()`**, but whitespace is the default character that is removed*.

### Working with the individual characters in strings

We know how to work with an entire string using some of the methods discussed, but what if we wanted to work with the individual characters? 

There are a couple of ways to do this, but first we'll focus on **indexing**. We know that to the Python compiler, a string is just an encoded collection of characters. It turns out that we can access the individual characters simply by asking Python for a given numbered element in the collection (i.e. the string). We do this by placing the element number that we want in squared brackets, `[]`, after our string (or named variable, if the string is stored in one). This element number is referred to as the **index** of the character (or **element**, if it's not a string - more on this later).

````python
my_str_variable = "Test String"
my_str_variable[1]
````

````python
my_str_variable[5]
````

````python
my_str_variable[-1]
````

````python
my_str_variable[-3]
````

It is possible to access any element of a string using an indices like this.

But why is the element at index 1 `e` and not a `T`? After all, `T` is the first element in the string? Also, what are those negative numbers doing? 

In the case of the former, it turns out that Python (and many other programming languages) is **zero indexed**, i.e. starts indexing at 0. 
This means that the first element in our string (and any collection that supports indexing) is accessed via indexing at 0. 
As for the negative numbers, this is a way to access elements backwards, starting from the end of the string rather than the beginning. Indexing from the end starts from -1 and continues downwards from there. So, we would use -2 to access the `n` in our "string".

You can also access any given number of characters (or **substring**) by combining multiple index numbers separated by a colon `:`. 

For example:

````python
my_str_variable[1:3]
````

````python
my_str_variable[5:9]
````

````python
my_str_variable[-6:-1]
````

````python
my_str_variable[1:]
````

````python
my_str_variable[:-1]
````

This indexing is very useful but can catch people out.

You might notice that when indexing our string variable using `[1:3]`, only the letters at index 1 and 2 are returned; when indexing from `[5:9]`, we get the letters at indices 5, 6, 7, and 8. Why?

This is because the indices that you pass in are **inclusive on the left side, and exclusive on the right side**. This means that when you index this way, you will **grab** letters from the ["start index":] that you give up to, but not including, letters at the given [:"end index"]. 

What about the last two examples, where there isn't an ending index or a starting index? 
If you don't give an ending index, Python assumes that your ending index is the last index in the string. 
Similarly, if you don't give a starting index, Python assumes that your starting index is the first index in the string. 
**Remember, this is the `zeroth` index in Python** (don't worry if this feels confusing, you'll get used to it quickly).

Furthermore, there is a way to grab elements at regular intervals in a string.
For example, what if we wanted to grab every second character? 
Python allows us to do this by passing in an optional third number while indexing. 

The optional third number, also separated by a colon (`:`), tells Python the **step size** by which to move through the string when indexing. So, if for whatever reason every second letter from the beginning to end is important, we could index it with `[::2]`. 

````python
my_str_variable[::2]
````

Or, if we wanted to grab every 3rd letter from index 2 up to, but not including, index 10, we could use the indexing `[2:10:3]`.

````python
my_str_variable[2:10:3]
````

### Iterating Through Strings

We can cycle through all of the letters in string (a process called **iteration**) in a couple of different ways. 

First, lets look at cycling through with a `while` loop.

````python
my_str = "HELLO"
i = 0  # Index
# While loops, loop until the conditional is fulfilled.
while i < 5:
    print(my_str[i])
    i += 1
````

The above "while" loop will **iterate** over the letters of our string `HELLO`, printing each one until the index `i` reaches the value 5. Since we knew the length of our string (i.e "HELLO" is 5 letters long), we knew that we could use the condition `while idx < 5:` for our loop checking, and ensure that all the letters would be printed due to **Zero Indexing**. 

What if we didn't know the length ahead of time? 
There is actually a string function that we can use to figure this out (we'll talk much more about functions and how they work in a later workbook). 

The function is `len()` for "Length", and we simply call `len()` with our string passed as an argument inside the funciton.
It returns the length of our string.

````python
len(my_str)  # my_string = "HELLO"
````

Now, we can write a more general `while` loop:

````python
while i < len(my_str):
    print(my_str[i])
    i += 1
````

As this is a Codealong, you will learn more by typing the code yourself. The next line is there for you to exactly do that and see the result.

````python
# Repeat the HELLO loop with len() so the code adapts to the string length.
my_str = "HELLO"
i = 0

while i < len(my_str):
    print(my_str[i])
    i += 1
````

Great! 

There are other ways to iterate over the letters in our string.

The other way that we can iterate over the letters in a string is to use a `for` loop. 

`for` loops are built off of the same idea as `while` loops (doing something over and over again), but instead of continuing until some condition is no longer met, `for` loops operate directly on iterables. 

This leaves the concern about when the loop stops and how does Python figure this out?
With a `for` loop, we don't have to care how many iterations/cycles the loop will go through. This comes with both its advantages and disadvantages adn will be discusses later.

Look at the syntax of a `for` loop:  

````python
for i in range(len(my_str)):
    print(my_str[i])
````

Try coding it yourself:

````python
# A for loop can iterate over index numbers too.
my_str = "HELLO"

for i in range(len(my_str)):
    print(my_str[i])
````

### Important Notes: 

The `range()` function (which we will cover in more depth when we get to the functions notebook) as used above simply gives us a list of numbers from 0 up to, but not including, the inputted number. 
Therefore, In the case above, since `len(my_str)` = 5, `range(len(my_str))` returns a list of integers from 0 to 4.

This `for` loop does the exact same thing as the `while` loop, but with slightly different syntax. 

How does it work? 

At each iteration of the loop,`i` is assigned one of the values in `range(len(my_str))`, and then the code within the indented block is run with that value of `i`. 

How does Python know what the values of `i` will be? Python simply goes through the values of whatever is after the `in` statement **in order**, and assigns those values to `i` one at a time through each iteration of the loop. 

Since `range(len(my_str))` returns to us a list of integers from 0 to 4, those values get assigned to `i` as we run through the `for` loop. Let's look at one of our favorite kinds of tables to view this:

| After loop # | idx | What's Printed |
| ------------ |:---:|:--------------:|
|      1       |  0  |       'h'      |
|      2       |  1  |       'e'      |
|      3       |  2  |       'l'      |
|      4       |  3  |       'l'      |  
|      5       |  4  |       'o'      |

With our `for` loop, the `i` variable is automatically changed, rather than to manually update it like we did in the `while` loop. 

This is one of the incredibly nice aspects of `for` loops! But wait, it gets even better!

It turns out that the above implementation of our `for` loop is actually considered to be **non-Pythonic**. This is because the way that `for` loops are constructed allows us to achieve the same output as above by writing the following:

````python
for c in my_str:
    # c = character
    print(c)
````

What's going on here!? 

Well, instead of iterating over all of the integers in a `range(len(my_str))` call like we did in our first `for` loop, we've made Python simply iterate over all of the individual characters in our string, `my_str`. 

In each iteration of this `for` loop, `c` stores a different letter of `my_str`, and then the call `print(c)` prints that character. In the end, we get the same result as either of our `while` loops, and the less Pythonic `for` loop, above. This way is considered to be the Pythonic way to iterate over a string, and so it's an important concept to grasp.

Why is it more Pythonic? 

When something is more "Pythonic", it means that the language is used in such a way that makes the code more readable, and simultaneously uses Python's power to make your solutions more optimal. Let's look at how this applies to the final implementation of our `for` loop.

We can see that it is more readable since we don't have to index into our string anymore. This means that there is less to follow along with and keep track of; rather than keeping track of both the current index we are on and what letter that index corresponds to in our string, all we have to keep track of is the current letter we're on. 

We can also note that our code just looks cleaner and simpler, too. In terms of making our code more optimal, since we no longer have to index into the string to grab characters, we have fewer steps in each iteration of the loop. This means less work for Python to do.

### A Quick Aside on String Formatting 

There's another thing that we should talk about before moving on from our discussion of strings - string formatting. 

String formatting is going to allow us to format strings in certain ways. Probably it is most useful for allow us to insert variable contents into strings dynamically. 

Let's look at the different syntax styles.  

```python
my_name = 'Sean'    # string variable 

print('Hello %s' % my_name)
print('Hello {}'.format(my_name))
print(f'Hello {my_name}')
```

The result is the same for each. See for yourself by typing the code in a python cell. 

````python
# Iterating directly over a string is the most readable option here.
my_str = "HELLO"

for character in my_str:
    print(character)
````

How is this working? 

Well, in each case, it's filling in a given part of our string with the value of the variable "my_name". 

In the first case, we use a `%` sign to denote where the replacement should happen, followed by a letter to denote what type of variable will be passed in there (`s` is used for string, `d` is for a decimal, etc.). You can find what each letter denotes [here](https://docs.python.org/2/library/stdtypes.html#string-formatting). 

In the second case, we use pointed brackets `{}` to denote where the replacement should take place. We can also place numbers, or even variable names themselves inside these brackets and reference them in the `format()` method or the f-string (`f"{variable}"`).

```python
print('Hello {0}'.format(my_name))
print('Hello {name}'.format(name=my_name))
```

This is something that we don't use much past pretty simple cases, but there are many more things you can do with it - you can read about them [Here](https://docs.python.org/2/library/string.html#format-specification-mini-language). In general, string formatting is much more readable and dynamic as compared to a bunch of concatenation.

Try coding some examples for yourself:

````python
# f-strings let us insert variable values into readable text.
name = "Ada"
favorite_language = "Python"

print(f"{name} is learning {favorite_language}.")
````
