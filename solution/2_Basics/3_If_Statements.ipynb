## The Logic

The simplest way to control the flow of a Python program is with an `if` statement. An `if` statement allows us to check if a certain condition is met. 

If the condition is `True`, certain operations will be executed. Otherwise, it is `False` and the operations will not be executed.

For example, say we're asked to write a program that takes a bunch of numbers and gives back to us the numbers that are even?

First, We would need to write an `if` statement that identifies whether or not a number is even (we'll talk later about how to do this).
Then, give back only those numbers that meet the "even" condition. This is a program that will be entirely within our ability to implement at the end of this repository; for now let's focus on the `if` statement.

The general syntax of a Python `if` statement is:

```python
if condition:
    if_block_statement # Indented
```

Notice how the `if` statement ends in a colon `:` after the `condition`. This is the way that Python declares the start of an indentation block. The purposes of indentation blocks manifest themselves in many different ways. With an `if` statement, just know that they mark a section of code that is run when the condition (before the colon) is True. 

What does it mean for a condition to be True? To understand this better, we will now look at conditionals.

### **Elif** and **Else**

The `elif` statement is just like the `if` - it accepts a condition to check the condition and has an indented code block that is executed when that condition evaluates to `True`. 

The `else` is similar, but it doesn't accept a condition. Instead, it acts as a catch all for any other situation that you didn't cover with your `if` and `elif` statements.

**Note**: In an `if`-`elif`-`else` block, there can only be a single `if` and a single `else`, but any number of `elif` statements. L

Let's take a closer look at this in the following code block:

````python
# Sample integer value for a runnable input example.
x = -3

# if-elif-else block
if x < 0:
    print("You entered a negative number.")
elif x > 0:
    print("You entered a positive number.")
else:
    print("You entered the number 0.")
````

Run the above cell and enter different numbers. 

Before you submit the number, think about what you think will be printed based on what you see in the conditionals. 

Can you understand what this `if`-`elif`-`else` block is doing?

Let's specifically talk about how the `if`-`elif`-`else` statement works. The programmers of Python designed these statements so that they would execute with high efficiency. 

To do this, Python goes through the `if`-`elif`-`else` statements one at a time. When it encounters a condition that evaluates to `True`, it will execute the corresponding conditional code block **and** then skips to the line directly following the last conditional block. 

Let's examine this in the following code:

````python
x = 7

if x > 5:
    print("You entered a number greater than 5.")
elif x > 0:
    print("You entered a positive number.")
elif x < 0:
    print("You entered a negative number.")
else:
    print("You entered the number 0.")
````

Try running the above code and enter the number 5. Then run it again and enter a number greater than 5 but before you do so, what do you think will be printed? 

Try it out. You may get slightly unexpected results but they will soon make sense.

Knowing what is going on in the above code block will give you full control over the flow of your programs. 

In the first case, when you entered 5, you got something unsurprising. The only condition that evaluates to true when `x` is 5 is the second one, **""You entered a positive number""**.

However, the second example prints only "You entered a number greater than 5." But a number greater than 5 is also greater than 0 so why did it not also print **'You entered a positive number'**?

This shows that only one of the conditional blocks in the `if`-`elif`-`else` statement was evaluated. Sounds efficient, doesn't it?
It is because the order of these blocks matters. Python checks each condition from top to bottom, executing the code for the first True condition it encounters and skips the rest. Therefore, it is important to place the most specific conditions first, followed by more general ones.
```python
if condition_1:
    # Executes if condition_1 is True
elif condition_2:
    # Executes if condition_1 is False and condition_2 is True
else:
    # Executes if all previous conditions are False
```
**Note**: In this case, the `else` part of the code block is actually optional. If it is not included, then we'd notice that at least one of the conditional blocks in the `if`-`elif` statement will be evaluated.

Lets change the code block slightly to allow for all our logic to be run. 

````python
x = 7

if x > 0:
    print("You entered a positive number.")
    if x > 5:
        print("You entered a number greater than 5.")
elif x < 0:
    print("You entered a negative number.")
else:
    print("You entered the number 0.")
````

by **Nesting** an additional "if" statement inside your first "if" block you can now access further boolean logic.

Lets look at another example. 

````python
x = 3

if x > 0:
    print("You entered a positive number.")
    if x > 5:
        print("You entered a number greater than 5.")
    else:
        print("You entered a number less than 5.")
elif x < 0:
    print("You entered a negative number.")
    if x < -5:
        print("You entered a number lower than -5.")
else:
    print("You entered the number 0.")
````

**Nested** "if" statements in Python allow you to place one if statement inside another, enabling you to check **Multiple Conditions**. This structure is useful when you want to evaluate additional conditions after the first one is True.
```python
if condition_1:
    # Code to execute if condition_1 == true
    if condition_2:
        # Code to execute if condition_2 == True
    else:
        # Code to execute if condition_2 == False
else:
    # Code to execute if condition_1 == false
```

### And, Or & Not

There are plenty of times when we want to execute some specific code when more than one condition is True. 

Check out the following code snippet where we want to check if a number is between 5 and 10. 

````python
x = 7
if x > 5:
    if x < 10:
        print(x)
````

We can see that what this **Nested**-`if` statement is checking is if `x` is greater than 5, and if so, then the conditional block is entered. 

Next, it checks if `x` is less than 10, and if this is True `x` is printed. We can intuitively guess that there is a better way to check for this condition... And there is!

Python gives us full access to what are known as boolean operations. The ones that we will use most often are `and`, `or`, and `not`. Both `and` & `or` take two conditions as inputs, while `not` takes only a single condition. All of these operations return a single boolean. `

**The Logic:**

**and**: Requires both conditions to be `True` to return `True`, otherwise it will return `False` 

**or**: Requires only one of the conditions to be `True` to return `True`; the only time it returns `False` is when both inputs are `False` 

**not**: Switches the truth of the input condition. These operations are derived from formal logic, and you can find a full discussion [Here](https://en.wikipedia.org/wiki/Truth_table).

What this means is that we now have a natural language way to combine conditions. The previously nested `if` statement can now be written as a simple `if x > 5 and x < 10`. We can also chain other interesting conditionals together. Change the value of `x` in the next cell so that it isn't printed and in the cell after so that it is.

````python
x = 12
if x > 10 or x < 5:
    print(x)
````

````python
x = 12
if not (x <= 10 and x >= 5):
    print(x)
````

Notice how the first `if` in the above code snippet uses `or` logic, printing `x` if it is greater than 10 **`or`** less than 5. 

Inherently, this statement is also saying that it will print `x` if `x` is **"not"** between 5 and 10, which is expressed in the second `if` statement. 

This illustrates an important point - there are always more than one way to accomplish the same thing in programming.

You can also combine several statements, look at the below code for example:

````python
country = "Germany"
age = 18

if (country == "Germany" and age >= 18) or (country != "Germany" and age >= 18):
    print("This person can legally drive a car.")
else:
    print("You are either too young to drive or not in Germany!")
````
