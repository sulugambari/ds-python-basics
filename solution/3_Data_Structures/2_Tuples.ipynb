### A Matter of Mutability 

One thing that will come up as an important distinction in the structures is the concept of mutability. 

**Mutability** refers to the capability of an object to be changed after it has been instantiated. 

For example, with lists, we can change the contents at any arbitrary index and even grow the list dynamically... Lists are **mutable**

````python
sample_list = list(range(10))
sample_list  # This calls the variable to print it out.
````

````python
sample_list[4] = 0  # Changes the value at index 4 to zero
sample_list
````

````python
sample_list.append(1)  # Appends (adds) the value 1 to the end of the list
sample_list
````

Mutability is nice, but there are times when we do not want our data structure to be mutable. 

For example, if we're allowing a user of our program to have access to a data structure, one way to ensure that they cant change it (sometimes users may do this out of malice, and we want to try and prevent that) is to make the structure **immutable**. There are many more reasons why immutability is a desired trait but this is the obvious one.

You already know about some object/data types in Python. The first types are various numerics (`int`, `float`, `complex`) - these are all immutable. What! 

Immutable you say? But, I can change a value in a variable after it's been declared. True, but consider the following:

````python
# First declaration of x
x = 1
````

````python
# Change the value of x
x = 2
````

````python
x
````

How can numerics be immutable while at the same time allowing us to change the value of a variable that holds a numeric? 

What's going on under the hood when you assign a value to a variable is that Python puts that value or data structure into memory, and then simply associates the variable name with that value or data structure. Changing a variable, then, simply amounts to associating that name with a different object **"in memory"**. So technically, the declared **variable** is mutable but the associated objects contained in that named variable can be mutable or immutable. 

Using this same logic, we can see that strings are also an immutable object type. The contents of a string are put into memory and the variable name you have declared is associated with that string. When you want to change the variable to a different string, Python simply associates that name with a different string, which is also immutable. 

Lists are mutable! This means that you can change the structure of the list in addition to the names of the things that are in the list 

**notice** the specific use of names there; we'll come back to that in the next section.

## Tuples

Tuples are simply an immutable `list`. They are immutable, ordered collections of data. This means that once a tuple is instantiated, all you can do is access its contents. 

You cannot make a tuple longer. 

You cannot reassign what is in a tuple (there are some subtleties to this which we will discuss shortly). 

Similar to lists, tuples are declared by:

1. Passing an iterable to the `tuple()` constructor
2. With curly brackets (parenthesis), similar to creating lists
3. Without parenthesis (This works because Python automatically interprets comma separated **things** as tuples, unless explicitely stated otherwise).

````python
my_empty_tuple = ()  # This is how you create an empty tuple.
````

````python
my_first_tuple = tuple(
    [1, 2, 3]
)  # This is how you create a tuple from an iterable list.
my_first_tuple
````

````python
my_second_tuple = tuple(
    "string"
)  # This is how you create a tuple from another iterable.
my_second_tuple
````

````python
my_other_tuple = (1, 2, 3)  # Seperated values with commas in parenthasis == tuple
my_other_tuple
````

````python
my_last_tuple = (
    1,
    2,
    4,
    "four",
)  # You can also create a tuple without parenthasis, just seperate values by commas.
my_last_tuple
````

````python
print(type(my_empty_tuple))
print(type(my_first_tuple))
print(type(my_second_tuple))
print(type(my_other_tuple))
print(type(my_last_tuple))
````

You may ask, What are the direct implications of using a tuple versus a list? 

Well, suppose you are trying to grab all even numbers stored in some collection of numbers, say the numbers 1-20. If we were to do this with a list, that might look like the following...

````python
evens = []
for element in range(1, 21):
    if element % 2 == 0:
        evens.append(element)
evens
````

You could try to do this using a tuple instead of a list by changing `evens = ()`, but once you try to run the code you would immediately get an error that says:
```bash
AttributeError: 'tuple' object has no attribute 'append'
```
Try this out by changing the code in the cell above. 

The error message is self explanatory. Again, error messages are our friends. It tells us that tuples have no ability to append, i.e. no `append()` method, and this is just as expected given that they are immutable.

You might be asking yourself what a tuple can/should store? The answer is, just as with lists, anything! Just as with lists, the elements of tuples can be accessed via zero-based indexing, and looped through with a `for` loop. And just as with lists, the elements in a tuple can be either homogeneous or heterogeneous (There are structures in Python that enforce homogeneity, we will cover that later.). 

Let's stick with looking at tuples for now, and take a look at some of the things we can store in them.

````python
t = (1, "bird", [1, 2], (3, 4), 5.0)
t
````

````python
print(type(t[0]))
print(type(t[1]))
print(type(t[2]))
print(type(t[3]))
print(type(t[4]))
````

The tricky thing about tuples is that even though they are immutable, if they are storing any mutable data types, those structures **can** be changed!

````python
t[2].append(3)
t
````

**Note**: The `append()` method is used directly on something that doesn't look like a list. why? This works because Python, upon accessing the contents of the tuple `t` at index 2, finds a list. It can then call the `append()` method directly on that data type.

A final thing to note is that since tuples are immutable, they have only two methods associated with them:

1. `count()` 
2. `index()`

If you cant guess what they do then try them out youself on your tuple `t`.

For this reason, we say that tuples are very lightweight; they don't take up much space in memory, but also don't have much built in functionality.

````python
# Part 2: tuples are immutable, but mutable objects inside them can change.
my_tuple = (1, "hello")
print(my_tuple[0])
print(my_tuple[1])
print(my_tuple[1][1])

print("You cannot assign my_tuple[1] = 'hello there' because tuples are immutable.")

my_other_tuple = ("other", [])
my_other_tuple[1].append("there")
my_other_tuple[1].insert(0, "hello")
print(my_other_tuple)
print("The tuple still contains the same list object; only the list contents changed.")
````
