## Working with Individual Elements in Lists

In python, working with individual elements in a list is the same as working with elements in strings. We can access the elements by indexing. 

````python
my_lst = [1, 2, "hello", "goodbye"]
````

````python
# indexing
my_lst[1]
````

````python
# slicing
my_lst[2:3]
````

**Remember**: The end index is non-inclusive.



````python
my_lst[:]
````

````python
my_lst[-1]
````

Just as with strings, we can also add a 3rd number to our indexing, allowing us to step through the list and only grab elements at regular intervals.

````python
my_lst = ["Lists", 1, 2, "Are", 4, 5, "Pretty", 7, 8, "Cool", 10]
````

````python
# Grab every third element of whole list
my_lst[::3]
````

````python
# Grab every third element ending with element #4
my_lst[:4:3]
````

````python
# Grab every third element starting with element #6
my_lst[6::3]
````

## Iteration

As you may have already guessed, we can also iterate through lists in the same way that we can iterate through strings. 

The most effective way to iterate through a list is to use a `for` loop. We could loop through the list with a `while` loop, but those are less Pythonic for lists and so a `for` loop is prefered.

````python
my_lst = [1, 2, 3, 4, 5]
````

````python
for num in my_lst:
    print(num)
````

````python
for num in my_lst:
    print(num + 4)
````

Just like iterating through strings, the `for` loop iterates over all of the characters in the `list`, and places those values into the declared variable name (`num`) at each iteration of the loop.

What if you absolutely need the indices? Is there a way that we can still iterate through using a "for" loop without `range()` and `len()`?

Yes! There is a function, `enumerate()`, that allows us to iterate through a list or string (grabbing each of the individual elements in the list or characters in the string) while at the same time keeping track of their index. 

The trick is that instead of using just one variable (such as `num`) to store the elements of the list as you loop through them, we use two variables. The first of these variables stores the index, and the second stores the corresponding element in the list.

Let's see how it works...

````python
my_lst = [1, 2, 3, 4, 5]

for i, num in enumerate(my_lst):  # i is the index, num is the integer at that index
    print(i, num)
````

````python
my_lst = [1, 2, "Three is a Magic Number", 4, 5]

for i, c in enumerate(my_lst):
    if i == 2:  # i is the index, c is the element at that index
        print(f"{c} at index {i}")
    else:
        print(f"Boring Number {c} at index {i}")
````

When `enumerate()` is called on our list, `enumerate()` gives us back two values at each iteration through the loop.

- The first value is the current index (which we chose to store as `i` above)
- The second value is the current element of the list (which we chose to store as `num`). 

Notice how `i` tracks one behind `num`... this is because... `i` starts at 0 and `num` starts at 1.


## The Magic of Iterables

Strings and lists are two kinds of iterables that we have already seen. There are many other kinds of iterables that you will work with in Python, and if you'd like you can even define your own iterable. 

The **Most Important** thing that we want to note for now is that the `for` loops that we have looked at today will work for **any iterable**. You can simply write `for <variable name> in <iterable>:`, and at each iteration through the loop you will be able to access another element from that iterable via `variable name`. 

Also, as mentioned before, the `list()` constructor accepts **any iterable** as an argument and then creates a `list` where each element in the iterable is a single element in the `list`.

````python
# Part 4: list iteration.
numbers = list(range(11))

for number in numbers:
    print(number)

evens = []
for number in numbers:
    if number % 2 == 0:
        print(number)
        evens.append(number)
print("Evens:", evens)

for number in numbers:
    if number % 2 != 0:
        print(number)

for index, number in enumerate(numbers):
    print(index, number)
````
