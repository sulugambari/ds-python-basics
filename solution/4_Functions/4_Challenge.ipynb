For the first part of the assignment, we're going to get some practice taking something we've already written and translating it to a function.

Write a function that determines whether or not an inputted number is prime, and then prints 'The number you inputted is a prime/ not a prime number.' depending on what your script finds (note that this means putting a `print` in front of the function when testing it will be redundant for this case). 

```python
# Part 1 solutions.


def factorial(number):
    """Return the factorial of a non-negative integer."""
    if number < 0:
        raise ValueError("factorial is only defined for non-negative integers")

    result = 1
    for factor in range(2, number + 1):
        result *= factor
    return result


for value in [0, 1, 3, 5, 7]:
    print(value, factorial(value))


def is_prime(number):
    """Print whether number is prime and return the boolean result."""
    if number < 2:
        print("The number you inputted is not a prime number.")
        return False

    for divisor in range(2, int(number**0.5) + 1):
        if number % divisor == 0:
            print("The number you inputted is not a prime number.")
            return False

    print("The number you inputted is a prime number.")
    return True


for value in [1, 2, 3, 9, 17]:
    is_prime(value)


# Part 2 solutions.


def count_words(text):
    """Count words separated by spaces."""
    return len(text.split())


def count_words_by_delimiter(text, delimiter=" "):
    """Count words separated by a chosen delimiter."""
    return len([part for part in text.split(delimiter) if part])


def word_lengths(text, delimiter=" "):
    """Return the length of each word-like part in a string."""
    return [len(part) for part in text.split(delimiter) if part]


def primes_up_to(number):
    """Return all prime numbers up to and including number."""
    primes = []
    for candidate in range(2, number + 1):
        if all(
            candidate % divisor != 0 for divisor in range(2, int(candidate**0.5) + 1)
        ):
            primes.append(candidate)
    return primes


def divisibility_labels(numbers, divisor):
    """Return yes/no labels for divisibility by divisor."""
    return ["yes" if number % divisor == 0 else "no" for number in numbers]


def strings_ending_with(strings, letter):
    """Return strings that end with the selected letter."""
    return [text for text in strings if text.endswith(letter)]


def indices_containing(strings, substring):
    """Return indices of strings that contain substring."""
    return [index for index, text in enumerate(strings) if substring in text]


print(count_words("This is a test string"))
print(count_words_by_delimiter("red,green,blue", ","))
print(word_lengths("This is a test string"))
print(primes_up_to(20))
print(divisibility_labels([10, 25, 36, 12, 20], 5))
print(strings_ending_with(["I", "am", "in", "love", "with", "Python"], "n"))
print(indices_containing(["This", "is", "an", "example"], "is"))


# Extra challenge solution.


def calculate_tax(tax_brackets, income):
    """Calculate tax owed for sorted or unsorted progressive tax brackets."""
    sorted_brackets = sorted(tax_brackets, key=lambda bracket: bracket[0])
    tax_owed = 0
    lower_bound = 0

    for upper_bound, rate in sorted_brackets:
        taxable_amount = min(income, upper_bound) - lower_bound
        if taxable_amount > 0:
            tax_owed += taxable_amount * rate
            lower_bound = upper_bound
        if income <= upper_bound:
            return tax_owed

    # Income above the final bound is taxed at the final bracket's rate.
    final_rate = sorted_brackets[-1][1]
    tax_owed += (income - lower_bound) * final_rate
    return tax_owed


brackets = [(50000, 0.08), (100000, 0.10), (150000, 0.15)]
unsorted_brackets = [(150000, 0.15), (50000, 0.08), (100000, 0.10)]

print(calculate_tax(brackets, 70000))
print(calculate_tax(brackets, 170000))
print(calculate_tax(unsorted_brackets, 170000))
```
