**<center>Device platform views (%)<center>**

```python
import matplotlib.pyplot as plt

device = ["TV", "Phone", "Tablet"]
views = [60, 30, 10]

numb_bins = 3
plt.bar(device, views)
plt.title("Percentage Views by Device")
plt.xlabel("Device")
plt.ylabel("Views")
plt.show()
```
