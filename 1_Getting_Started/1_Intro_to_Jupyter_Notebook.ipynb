## Open a Notebook and Select a Kernel

VS Code can open existing `.ipynb` files directly from the Explorer. You can also create a new notebook with the command palette command `Create: New Jupyter Notebook`.

When you open a notebook, select a kernel from the kernel picker in the top-right corner. The kernel is the Python environment that runs your notebook code. For this repository, choose the environment created by `uv sync`.

## Running Cells

**Welcome to programming!**

A notebook is made of cells. A cell can contain Python code, Markdown notes, or other content.

To run a code cell in VS Code, select it and press **Shift+Enter**. You can also click the **Run** icon next to the cell. The output appears directly below the cell.

Try running the code cell below:

```python
print("This is the code being run. Here is the output of the code block.")
```

The output of the code appears directly beneath the cell.

VS Code also supports other run actions. You can run the selected cell, run all cells, or run cells above/below the current cell from the notebook toolbar.

## Cell Types

A notebook usually contains two important cell types:

- **Code cells**, where you write and run Python code
- **Markdown cells**, where you write notes, headings, explanations, and links

In VS Code, select a cell and use the language picker in the bottom-right corner of the cell to switch between code and Markdown. You can also use keyboard shortcuts in command mode: **M** changes the selected cell to Markdown, and **Y** changes it to code.

To render a Markdown cell, use the check mark in the cell toolbar or run the cell with **Shift+Enter**.

![Change language](../assets/change-language.png)

![Rendered Markdown displayed in code cell](../assets/markdown-rendered.png)

## Editing and Managing Cells

Most notebook actions are available from the cell toolbar, keyboard shortcuts, or the command palette.

Common actions:

- Add a new code or Markdown cell
- Move cells up or down
- Copy, cut, and paste cells
- Delete a selected cell
- Edit a Markdown cell by double-clicking it
- Run a selected cell with **Shift+Enter**

If a shortcut behaves differently on your computer, open the command palette with **Cmd+Shift+P** on macOS or **Ctrl+Shift+P** on Windows/Linux, then search for the notebook action by name.

## Running Terminal Commands

You can run terminal commands directly from a notebook cell by starting the line with `!`.

**Note:** terminal commands should be kept in their own cells. Do not mix them into the same cell as normal Python code.

Try it out:

```python
!pwd
```

```python
!ls
```

Start typing a name, method, or function and VS Code will suggest completions. You can usually accept a suggestion with **Tab** or **Enter**. You can also hover over functions and variables to see quick information when VS Code can infer it.

## Stop, Restart, and Clear Outputs

Long-running Python processes can be interrupted. In VS Code notebooks, use the notebook toolbar to interrupt or restart the kernel, clear outputs, or run all cells.

Restarting the kernel clears the current Python session. This is useful when you want to check that the notebook still works from top to bottom.

![Notebook Toolbar](../assets/notebook-toolbar.png)

Try running the next cell and stopping it if needed:

```python
import time

print("Sleeping briefly")
time.sleep(1)  # Short sleep so the notebook remains runnable.
print("Done Sleeping")
```
